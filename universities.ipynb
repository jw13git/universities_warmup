```python
type_uni = uni_df['Type']
type_uni = type_uni.str.strip()
type_uni = type_uni.str.split(',', expand=True)
type_uni
```

```python
associates = []
for i in type_uni[1]:
    if 'primarily' in i:
        associates.append(1)
    elif 'primarily' not in i:
        associates.append(0)
    else:
        associates.append(0)
```

```python
associates = pd.Series(associates)
associates
```

```python
uni_df['associates'] = associates
uni_df
```

```python
Profit_Non_Profit = []
for i in uni_df['Type']:
    if 'not-for' in i:
        Profit_Non_Profit.append('Not-For-Profit')
    elif 'Public' in i:
        Profit_Non_Profit.append('Not-For-Profit')
    else:
        Profit_Non_Profit.append('For-Profit')
# associates = []
# for i in type_uni[1]:
#     if 'primarily' in i:
#         associates.append(1)
#     elif 'primarily' not in i:
#         associates.append(0)
#     else:
#         associates.append(0)
# associates
```
